```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

orig = plt.imread('nick.png')[:, :, :3]
im = orig.copy()
im = colors.rgb_to_hsv(im)
wall_hue = im[100, 100, 0]
thresh = .05
wall_area_mask = (im[:, :, 0] > wall_hue - thresh) & (im[:, :, 0] < wall_hue + thresh)
im[wall_area_mask, 0] = .92
im[wall_area_mask, 1] *= .6
# im[:, :, 0] = 1.
# im[:, :, 1] = 1.
# im[:, :, 2] = 1.
im = colors.hsv_to_rgb(im)

fig, axes = plt.subplots(ncols=2)
axes[0].imshow(orig);
axes[0].set_title('Original Image')
axes[1].imshow(im);
axes[1].set_title('New Hue');
```
