# Intro to GitHub

### Profile Settings

There are two levels of settings in GitHub:
  - **Repo Settings**: Every repository can have its own settings; for example, whether it's public or private, who is allowed to contribute, whether to delete the repo, etc.  Repo settings are found under the "Settings" tab.

![repo settings button](imgs/github_reposettings.png)


  - **Account Settings**.  Your account has settings, too; a lot of them!  These affect how you appear on GitHub and how it appears to you and how you log in and connect your projects to it from your computer.  Looking at your account settings doesn't take long, and it's worth it because it helps improve your experience with GitHub. Account settings are found by clicking on your profile icon in the upper-right corner of the screen, then clicking "Settings" in the dropdown box that appears.
  
![account settings button](imgs/github_accountsettings.png)

### Register an SSH Key with Your GitHub Account

SSH keys are an alternative to passwords for authenticating interactions; instead of giving a password each time you want to do something, you just send an SSH key.  The neat thing about SSH keys is that they are specific to the machines that created them; only the computers that you've registered can do the SSH-specific actions.  

Because of this, this step may be needed when using your company's private GitHub, to meet their security needs. Without using SSH, you might see "permission denies" or similar types of errors when using your company's GitHub.

![ssh settings](imgs/github_ssh.png)

Registering an SSH key involves three steps:
  1. Generating a matching pair of keys (a "keypair") on your computer, usually in the terminal with the `ssh-keygen` program.
    - Open a Terminal
      - In Windows you'll need Git installed, then open the Git Bash program.
      - In Mac or Linux just open any terminal
    - [ ] type `ssh-keygen` and press enter.  This generates a pair of keys.  It will ask if you want to add a password to them.  You can, it will add even more security!  Hit enter until the program says it made the files
  2. Viewing the Public key text from the file that's generated (usually in a text editor, but we'll show how to use the `cat` program to print it in the terminal)
    - The file where the public key is found after where it says `Your public key has been saved in `.  
    - [ ] Copy-paste that filepath into the terminal, making the following command: `cat /path/to/my/file.pub`
    - `cat` is just a program that prints the text contents of any file.  If you type `cat` and then press enter, you might find you can't edit the terminal.  Just hit `Ctrl-Enter` on the keyboard and the program will stop so you can interact with the terminal again.
  3. Copy-pasting the Public key text into GitHub in the "New SSH Key' menu, and giving the key a name to help you remember what computer it's linked to.
    - [ ]  In the "SSH and GPG keys" menu, click the "New SSH Key" button and paste the key into the text box, with any title you choose.  Submit it with "Add SSH Key".  If GitHub accepts it, you should be done!  But how can you be sure?
  4. [ ] You can [test](https://docs.github.com/en/authentication/connecting-to-github-with-ssh/testing-your-ssh-connection) that the SSH connection works from the command line with the following terminal command: `ssh -T git@github.com`.  If it works, the first time it will say "The authenticity...cannot be established.  Are you sure? (yes/no)" Type yes (this message appears a lot in SSH connections), then enter, and you should get a success message!

We won't go into how to use ssh with Git yet, but walking through this step will help you orient to using GitHub and make it easier for you to get over the security hurdles when first using GitHub at your company.
