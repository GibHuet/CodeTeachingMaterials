# If-else Blocks
## Python's Block Syntax

If-else blocks are used in programming languages to control the "flow" of the program (i.e. make decisions about which code gets run).  

```python
x = 5
if x > 4:
    print('big')
elif x < 0:  # optional, can have as many as you want
    print('negative')
else:   # optional 
    print('small')
```
